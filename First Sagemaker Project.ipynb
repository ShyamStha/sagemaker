## Train Model

```python
import sagemaker

region = sagemaker.Session().boto_region_name
print("AWS Region: {}".format(region))

role = sagemaker.get_execution_role()
print("Rolearn: {}".format(role))
```

```python
sagemaker.__version__
```

```python
from sagemaker.debugger import Rule, rule_configs
from sagemaker.session import TrainingInput

s3_output_location='s3://{}/{}/{}'.format(bucket, prefix, 'xgboost_model')
container= sagemaker.image_uris.retrieve("xgboost", region, "1.2-1")
print(container)

xgb_model=sagemaker.estimator.Estimator(
    image_uri=container,
    role=role,
    instance_count=1,
    instance_type='ml.m4.xlarge',
    volume_size=5,
    output_path=s3_output_location,
    sagemaker_session=sagemaker.Session(),
    rules=[Rule.sagemaker(rule_configs.create_xgboost_report())]
)
```

```python
xgb_model.set_hyperparameters(
    max_depth = 5,
    eta = 0.2,
    gamma = 4,
    min_child_weight = 6,
    subsample = 0.7,
    objective = "binary:logistic",
    num_round = 1000
)
```

```python
from sagemaker.session import TrainingInput

train_input = TrainingInput(
    "s3://{}/{}/{}".format(bucket, prefix, "data/train.csv"), content_type="csv"
)
validation_input = TrainingInput(
    "s3://{}/{}/{}".format(bucket, prefix, "data/validation.csv"), content_type="csv"
)
```

```python
xgb_model.fit({"train": train_input, "validation": validation_input}, wait=True)
```

## Deploy the model

```python
import sagemaker
from sagemaker.serializers import CSVSerializer
xgb_predictor=xgb_model.deploy(
    initial_instance_count=1,
    instance_type='ml.t2.medium',
    serializer=CSVSerializer()
)
```

## Evaluate the Model

```python
import numpy as np
def predict(data, rows=1000):
    split_array = np.array_split(data, int(data.shape[0] / float(rows) + 1))
    predictions = []

    for array in split_array:
        # Returns bytes, decode first
        pred_bytes = xgb_predictor.predict(array)
        pred_str = pred_bytes.decode('utf-8')
        
        # Convert CSV string to list of floats
        pred_list = [float(x) for x in pred_str.strip().split(',')]
        predictions.extend(pred_list)

    return np.array(predictions)
```

```python
import matplotlib.pyplot as plt

predictions=predict(test.to_numpy()[:,1:])
plt.hist(predictions)
plt.show()
```

```python
import sklearn

cutoff=0.5
print(sklearn.metrics.confusion_matrix(test.iloc[:,0], np.where(predictions > cutoff, 1, 0)))
print(sklearn.metrics.classification_report(test.iloc[:,0], np.where(predictions > cutoff, 1, 0)))
```

```python
import matplotlib.pyplot as plt

cutoffs = np.arange(0.01, 1, 0.01)
log_loss = []
for c in cutoffs:
    log_loss.append(
        sklearn.metrics.log_loss(test.iloc[:,0], np.where(predictions > c, 1, 0))
    )

plt.figure(figsize=(15, 10))
plt.plot(cutoffs, log_loss)
plt.xlabel("Cutoff")
plt.ylabel("Log loss")
plt.show()
```

```python
print(
    'Log loss is minimized at a cutoff of ', cutoffs[np.argmin(log_loss)],
    ', and the log loss value at the minimum is ', np.min(log_loss)
)
```

```python
import numpy as np
from sagemaker.predictor import Predictor

# Deployed endpoint
predictor = Predictor(endpoint_name="sagemaker-xgboost-2025-10-21-05-17-31-187")

# Example: 1 row of features from your data (exclude target Income>50K)
features = np.array([[62.0, 6, 4.0, 6, 8, 0, 4, 0, 0.0, 0.0, 66.0, 39]])

# Convert to CSV bytes (SageMaker expects bytes for the Body)
csv_input = "\n".join([",".join(map(str, row)) for row in features])
csv_bytes = bytes(csv_input, encoding="utf-8")

# Make prediction
predictions_bytes = predictor.predict(csv_bytes, initial_args={"ContentType": "text/csv"})

# Convert result back to NumPy array
predictions = np.fromstring(predictions_bytes.decode("utf-8"), sep=",")
print(predictions)
```
